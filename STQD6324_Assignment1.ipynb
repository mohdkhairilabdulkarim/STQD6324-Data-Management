## **<font color='#da9100'>Aim 3:</font>** To find any pattern of food production in Malaysia.

```python
# Drop the 'Entity' column if it's not needed for trend analysis
df_trend_analysis = df_msia.drop(['Entity'], axis=1).set_index('Year')

# Create an empty dictionary to store trends
food_trends = {}

# Iterate through each column in df_trend_analysis
for food in df_trend_analysis.columns:
    # Use numpy's polyfit to fit a linear trend line to the production data over years
    # np.polyfit returns the coefficients of the polynomial, where the first coefficient is the slope
    slope, intercept = np.polyfit(df_trend_analysis.index.astype(float), df_trend_analysis[food], 1)
    food_trends[food] = slope

# Separating foods into upward and downward trends based on the slope
upward_trend = {food: slope for food, slope in food_trends.items() if slope > 0}
downward_trend = {food: slope for food, slope in food_trends.items() if slope < 0}

# Display the results
print("Foods with Upward Trend:", upward_trend)
print("Foods with Downward Trend:", downward_trend)
```

```python
# Get the top 2 foods with the strongest upward trends
top_two_upward = sorted(food_trends.items(), key=lambda x: x[1], reverse=True)[1:3] # Excluding Yams due to above issue.

# Get the 2 foods with the strongest downward trends
bottom_two_downward = sorted(food_trends.items(), key=lambda x: x[1])[:2]

# Extracting food names for plotting
foods_to_plot_upward = [food for food, _ in top_two_upward]
foods_to_plot_downward = [food for food, _ in bottom_two_downward]

# Plotting the production trends for these foods
plt.figure(figsize=(12, 6))

# Plot upward trends
for food in foods_to_plot_upward:
    plt.plot(df_msia['Year'], df_msia[food], marker='o', linestyle='-', label=f"Upward - {food} (Trend: {food_trends[food]:.2f})")

# Plot downward trends
for food in foods_to_plot_downward:
    plt.plot(df_msia['Year'], df_msia[food], marker='o', linestyle='--', label=f"Downward - {food} (Trend: {food_trends[food]:.2f})")

plt.title('Top 2 Upward and Bottom 2 Downward Production Trends in Malaysia (2017-2021)')
plt.xlabel('Year')
plt.ylabel('Production (tonnes)')
plt.legend(title='Foods')
plt.grid(True)
plt.show()
```

```python
# Calculate the regression line
slope, intercept = np.polyfit(df_msia['Maize'], df_msia['Sugar cane'], 1)
y_pred = slope * df_msia['Maize'] + intercept

# Calculate the error band (5% error)
error = 0.05 * y_pred
lower_bound = y_pred - error
upper_bound = y_pred + error

# Create a new DataFrame for plotting
plot_data = pd.DataFrame({
    'Maize': df_msia['Maize'],
    'Sugar Cane': df_msia['Sugar cane'],
    'Fit': y_pred,
    'Lower Bound': lower_bound,
    'Upper Bound': upper_bound
})

# Sorting data for band plotting
plot_data.sort_values('Maize', inplace=True)

# Using seaborn to plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Maize', y='Sugar Cane', data=plot_data, color='purple', label='Data Points')

# Plot regression line
plt.plot(plot_data['Maize'], plot_data['Fit'], color='black', label='Regression Line')

# Plotting the error band
plt.fill_between(plot_data['Maize'], plot_data['Lower Bound'], plot_data['Upper Bound'], color='gray', alpha=0.3, label='5% Error Band')

plt.title('Maize vs. Sugar Cane Production in Malaysia')
plt.xlabel('Maize Production (tonnes)')
plt.ylabel('Sugar Cane Production (tonnes)')
plt.legend()
plt.grid(True)
plt.show()
```
