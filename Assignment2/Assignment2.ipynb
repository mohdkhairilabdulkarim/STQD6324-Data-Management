```python
import pandas as pd

# Define the path to your Google Drive directory
data_path = '/content/drive/MyDrive/airline_data_results/'

# Load the data files
avg_delay_by_hour = pd.read_csv(data_path + 'avg_delay_by_hour/part-v025-o000-r-00000', header=None)
avg_delay_by_day = pd.read_csv(data_path + 'avg_delay_by_day/part-v011-o000-r-00000', header=None)
avg_delay_by_month = pd.read_csv(data_path + 'avg_delay_by_month/part-v023-o000-r-00000', header=None)
delay_factors = pd.read_csv(data_path + 'delay_factors/part-v009-o000-r-00000', header=None)
cancellations = pd.read_csv(data_path + 'cancellations/part-v020-o000-r-00000', header=None)
delays_by_flight = pd.read_csv(data_path + 'delays_by_flight/part-v016-o000-r-00000', header=None)
cancellations_by_flight = pd.read_csv(data_path + 'cancellations_by_flight/part-v008-o000-r-00000', header=None)
```

```python
# Column names for each DataFrame
avg_delay_by_hour.columns = ['Hour', 'AvgDelay']
avg_delay_by_day.columns = ['DayOfWeek', 'AvgDelay']
avg_delay_by_month.columns = ['Month', 'AvgDelay']
delay_factors.columns = ['AvgCarrierDelay', 'AvgWeatherDelay', 'AvgNASDelay', 'AvgSecurityDelay', 'AvgLateAircraftDelay', 'AvgTotalDelay']
cancellations.columns = ['CancellationCode', 'Count']
delays_by_flight.columns = ['FlightNum', 'DelayCount']
cancellations_by_flight.columns = ['FlightNum', 'CancelCount']
```

## **<font color=#FFD700>Findings</font>**

### **<font color=#4169E1>Objective 1:</font>** Optimal Times for Minimizing Flight Delays


1. **Average Delay by Hour of the Day**:


```python
avg_delay_by_hour
```

```python
import matplotlib.pyplot as plt

# Ensure the 'Hour' column is sorted
avg_delay_by_hour = avg_delay_by_hour.sort_values('Hour')

# Plot average delay by hour as a time series
plt.figure(figsize=(10,6))
plt.plot(avg_delay_by_hour['Hour'], avg_delay_by_hour['AvgDelay'], marker='o', linestyle='-')
plt.title('Average Delay by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Delay (minutes)')
plt.grid(True)
plt.show();
```

2. **Average Delay by Day of the Week**:


```python
avg_delay_by_day
```

```python
# Map numbers to day names
day_mapping = {1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday', 7: 'Sunday'}
avg_delay_by_day['DayName'] = avg_delay_by_day['DayOfWeek'].map(day_mapping)

# Ensure the days are ordered correctly and convert DayName to string
avg_delay_by_day['DayOfWeek'] = pd.Categorical(avg_delay_by_day['DayOfWeek'], categories=[1, 2, 3, 4, 5, 6, 7], ordered=True)
avg_delay_by_day = avg_delay_by_day.sort_values('DayOfWeek')
avg_delay_by_day['DayName'] = avg_delay_by_day['DayName'].astype(str)

# Plot average delay by day as a time series
plt.figure(figsize=(10,6))
plt.plot(avg_delay_by_day['DayName'], avg_delay_by_day['AvgDelay'], marker='o', linestyle='-', color='green')
plt.title('Average Delay by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Delay (minutes)')
plt.grid(True)
plt.show();
```

3. **Average Delay by Month**:


```python
avg_delay_by_month
```

```python
# Map numbers to month names
month_mapping = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}
avg_delay_by_month['MonthName'] = avg_delay_by_month['Month'].map(month_mapping)

# Ensure the months are ordered correctly and convert MonthName to string
avg_delay_by_month['Month'] = pd.Categorical(avg_delay_by_month['Month'], categories=list(month_mapping.keys()), ordered=True)
avg_delay_by_month = avg_delay_by_month.sort_values('Month')
avg_delay_by_month['MonthName'] = avg_delay_by_month['MonthName'].astype(str)

# Plot average delay by month as a time series
plt.figure(figsize=(10,6))
plt.plot(avg_delay_by_month['MonthName'], avg_delay_by_month['AvgDelay'], marker='o', linestyle='-', color='purple')
plt.title('Average Delay by Month')
plt.xlabel('Month')
plt.ylabel('Average Delay (minutes)')
plt.grid(True)
plt.show();
```

#### <font color=#4169E1>Insights</font>

1. **Average Delay by Hour of the Day**:
   - **Trend**: The average delay fluctuates significantly throughout the day.
   - **Early Morning Hours (12 AM to 6 AM)**: There is a significant drop in delays after the initial peak, with the lowest delays observed around 5 AM.
   - **Midday to Late Evening (7 AM to 11 PM)**: Delays gradually increase throughout the day, peaking significantly around 10 PM to 11 PM.

2. **Average Delay by Day of the Week**:
   - **Trend**: Delays vary across the days of the week, with noticeable peaks and troughs.
   - **Lowest Delays**: Wednesday and Saturday show the lowest average delays.
   - **Highest Delays**: Friday experiences the highest average delays, indicating it is the most problematic day for flight punctuality.

3. **Average Delay by Month**:
   - **Trend**: Delays show seasonal variations, peaking in February and decreasing significantly by April.
   - **Lowest Delays**: April has the lowest average delays.
   - **Highest Delays**: February experiences the highest average delays, possibly due to winter weather conditions and other seasonal factors.



#### <font color=#4169E1>Recommendations</font>

1. **Optimal Flight Scheduling**:
   - **Early Morning Flights**: Schedule flights to depart in the early morning hours, particularly around 5 AM, to minimize the risk of delays. This is the time when the average delay is the lowest.
   - **Midweek Flights**: Encourage travelers to book flights on Wednesdays and Saturdays to reduce the likelihood of delays, as these days show the lowest average delays.
   - **Seasonal Considerations**: Advise travelers to plan flights in months like April, when delays are typically lower, avoiding peak delay months such as February.

2. **Operational Improvements**:
   - **Peak Hour Management**: Implement measures to manage delays during peak hours, particularly in the late evening. This could include better coordination and scheduling practices, ensuring that ground operations are more efficient during these times.
   - **Day-Specific Strategies**: Develop strategies tailored to specific days of the week. For example, additional resources and contingency plans could be allocated for Fridays, which have the highest delays.

3. **Seasonal Adjustments**:
   - **Winter Preparedness**: Enhance operational preparedness and contingency planning for winter months, particularly in February, to mitigate the impact of weather-related delays.
   - **Spring Optimization**: Take advantage of the lower delay periods in spring (e.g., April) by scheduling more flights during these times, thus improving overall on-time performance.

4. **Passenger Communication**:
   - **Awareness Campaigns**: Conduct awareness campaigns to inform passengers about the optimal times and days for travel, helping them make better-informed decisions that align with lower delay periods.

```python
delay_factors
```

```python
# Plot delay factors
delay_factors.plot(kind='bar', figsize=(12,8))
plt.title('Average Delay by Factor')
plt.xlabel('Delay Factor')
plt.ylabel('Average Delay (minutes)')
plt.show();
```

#### <font color=#87CEEB>Insights</font>

1. **Average Delay by Factor**:
   - **Carrier Delay**: The average carrier delay is significant, suggesting that issues such as airline operations, staffing, and scheduling are major contributors to overall delays.
   - **Weather Delay**: Weather-related delays are relatively low compared to other factors, indicating that while weather can cause delays, it is not the most significant factor.
   - **NAS Delay (National Aviation System Delay)**: These delays are considerable, pointing to issues within the broader aviation infrastructure, including air traffic control and airport operations.
   - **Security Delay**: Security delays are minimal, indicating that security-related disruptions do not heavily impact overall delay times.
   - **Late Aircraft Delay**: This is the highest individual delay factor, showing that delays caused by late-arriving aircraft can significantly impact subsequent flights and overall scheduling.



#### <font color=#87CEEB>Recommendations</font>

1. **Improve Airline Operations**:
   - **Carrier Delay Mitigation**: Airlines should focus on optimizing their operations to reduce carrier-related delays. This could include better scheduling, increasing staffing levels during peak times, and improving maintenance processes to ensure aircraft are ready for departure on time.
   - **Predictive Maintenance**: Implement predictive maintenance technologies to foresee and address potential aircraft issues before they cause delays.

2. **Enhance Coordination with NAS**:
   - **Collaborative Decision Making**: Strengthen collaboration with air traffic control and airport operations to improve the efficiency of the National Aviation System. Implement technologies that enable better real-time decision making and communication between airlines and aviation authorities.
   - **Infrastructure Investments**: Advocate for investments in aviation infrastructure, including upgrades to air traffic control systems and airport facilities, to reduce NAS-related delays.

3. **Optimize Turnaround Times**:
   - **Reduce Late Aircraft Delays**: Focus on optimizing turnaround times for aircraft to minimize the cascading effect of delays. Implementing efficient boarding processes, timely baggage handling, and quick refueling can significantly reduce late aircraft delays.
   - **Buffer Scheduling**: Introduce buffer times in the scheduling of flights to account for potential delays, thus reducing the impact on subsequent flights.

4. **Leverage Technology**:
   - **Real-Time Data Analytics**: Utilize real-time data analytics to monitor and manage flight operations more effectively. This can help identify potential delay factors in advance and allow for proactive measures to be taken.
   - **Automated Systems**: Implement automated systems for scheduling and operations management to reduce human error and improve overall efficiency.

5. **Weather Contingency Planning**:
   - **Weather Monitoring Systems**: Invest in advanced weather monitoring systems to predict and mitigate weather-related delays more effectively.
   - **Flexible Scheduling**: Develop flexible scheduling strategies that can accommodate sudden changes in weather conditions without significantly disrupting operations.

6. **Passenger Communication**:
   - **Real-Time Updates**: Provide passengers with real-time updates about delays and expected departure times to manage expectations and reduce frustration.
   - **Customer Support**: Enhance customer support services to assist passengers during delays, offering rebooking options and accommodations if necessary.

By addressing these factors through targeted operational improvements and leveraging technology, airlines can significantly reduce delays, improve on-time performance, and enhance passenger satisfaction.
